```python
## first, a simple tree model will be used
## we want to see column attributes entropy distribution, before we use advanced fine tunning model, like boostring tree, CNN-VGG net

## ID3 purity measure, Ross Quinlan (ID3, 1986)
## C4.5, Ross Quinlan 1993
## Breimanetal(CART, 1984)

## here we use entroy to measure purity, e.g.:
##      attr (5 yes 1 no)
##      /  \
##    valA  valB
## (3 yes,  (2 yes, 1 no)
##  0 no)
## We say valA has higher purity
## To compute purity for attr, we evaluate it over all its sampling values

def entroy(p):
    # p = (klass_i / sample_p)
    if p >= 0 and p <= 1:
        return - np.log(p).dot(p)
    else:
        raise Exception("wrong value p! Does not comform with condition 0<= p <= 1")

def occur_counter(vec, objmap=None):
    if objmap != None:
        vec = objmap(vec)

    ret = {}
    for val in vec:
        if ret.get(val, None) is None:
            ret[val] = 0.0
        ret[val] += 1.0
    
    return ret / len(vec)
    
        
# Impure attr val will result in maximum of bits to tell wether it belongs to some class or not. The attr val gives you nothing
# using pandas selector will be easier
def val_entropy(attr_cliques, val):
    """
    mtx: c1 c2 c3
         ----------
                A  |
                A  |
         -----------
                B
                ...
    @attr_cliques: data set grouped by c3 values
    """
    _, sub_labels = attr_cliques[val]
    p = occur_counter(sub_labels)
    return entropy(p)

def all_entropy(clique):
    _, labels = clique
    p = occur_counter(labels)
    return entropy(p)

def attr_entropy(data, labels, attr):
    """
    @data: nunpy array
    @lables: numpy array
    """
    ## groupby data[:, attr]
    l = len(data)
    partition = {}
    # import itertools
    # for k, subset in itertools.groupby(data, key=lambda row: data[attr]):
    #     partition[k] = subset
    
    for i, row in enumerate(data):
        val = row[attr]
        if partition.get(val, None) is None:
            partition[val] = []
        partition[val].append(i)
        
    attr_cliques = {k: (data[subset_index,:], labels[subset_index]) for k, subset_index in partition.items()}
    ret = 0.0
    for k, sub_set in attr_cliques.items():
        sub_l = len(sub_set)
        ret += val_entropy(attr_cliques, k) * double(sub_l) / l
    return attr_cliques, ret
   
INF = -1e10
def get_best_split_fea(data, labels):
    ## compute total entroy
    H = all_entropy((data, labels))
    ## compute entory for each attribute
    max_gain = -INF
    # in a real application, we will do index converter here
    l = len(data)
    attrs = range(l)
    max_clique = None
    
    for i in attrs:
        attr_cliques, entropy = attr_entropy(data, labels, i)
        if max_gain < entropy:
            max_gain = entropy
            max_clique = attr_clique
    
    return max_clique, i
        

class jsonTree(object):pass
    
    
```

```python
CAFFE_ROOT = "../caffe/"

model_weights = os.path.join(CAFFE_ROOT, "lwf_caffe_face/face_model.caffemodel")
if os.path.isfile(model_weights):
    print("model %s is found!" % model_weights)
else:
    print("model file is not found!")
model_def = os.path.join(CAFFE_ROOT, "lwf_caffe_face/face_deploy.prototxt.txt")
if os.path.isfile(model_def):
    print("model_def %s is found!" % model_def)
else:
    print("model definition is not found!")

img = "0.jpg"
lfw_attr = "lfw_attributes.txt"

caffe.set_device(0)
caffe.set_mode_gpu()

net = caffe.Net(model_def,
                model_weights,
                caffe.TEST)

img=caffe.io.load_image(img)
target = img
test_img = caffe.io.load_image("Elizabeth_Smart_0004.jpg")

def detect(net, img):
    if img.ndim == 2:
        rows, cols = img.shape
    elif img.ndim == 3:
        rows, cols, ch = img.shape
    # compute scaling
    chunck_shape = (3,) +  net.blobs['data'].data.shape[1:]
    net.blobs['data'].reshape(*chunck_shape)
    net.reshape()
    # print("After reshaping ...")
    # print("net.blobs['data']", net.blobs['data'])
    # print("net.blobs['data'].data.shape", net.blobs['data'].data.shape)
    
    # print(net.blobs['data'].data.shape)
    transformer = caffe.io.Transformer({'data': net.blobs['data'].data.shape})
    transformer.set_transpose('data', (2,0,1))
    # transform mean
    
    # scale
    # transformer.set_raw_scale('data', 255)
    
    if img.ndim == 2:
        img = img[:,:,np.newaxis]
    # print("before transforming, image : ", img.shape)
    transformed_image = transformer.preprocess('data', img)
    # print("transformed image shape is : (%s,%s,%s)" % transformed_image.shape)
    net.blobs['data'].data[0] = transformed_image
    # Forward pass.
    out = net.forward()
    #
    return out
```

```python
import math
import numpy as np

def cosine_(v1, v2):
    v2.dot(v1)
    n1 = np.sqrt(v1.dot(v1))
    n2 = np.sqrt(v2.dot(v2))
    return v1.dot(v2) / (n1 * n2) 

def deepfeatureExtract(net, img):
    dim = [3, 112, 96]
    # coord5points = 
    # facial5points = 
    
    origin_img = img
    flip_img = np.flipud(img)
    
    out1 = detect(net, origin_img)['fc5']
    out2 = detect(net, flip_img)['fc5']
    
    return np.hstack([out1, out2])

plt.figure()
plt.imshow(test_img[70:185, 75:175])
target_fea = deepfeatureExtract(net, test_img[75:175, 75:175])
print(target_fea.shape)
```
